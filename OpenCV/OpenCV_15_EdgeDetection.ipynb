```
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cats.jpg')
img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/park.jpg')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()

# Labplacion - 초크, 펜슬로 그린 것 같은 느낌
lap = cv.Laplacian(gray, cv.CV_64F)
lap = np.uint8(np.absolute(lap))
plt.imshow(lap)
plt.show()

# Sobel - 깔끔하지는 않지만 많이 사용된다.
sobelx = cv.Sobel(gray, cv.CV_64F, 1, 0)
sobely = cv.Sobel(gray, cv.CV_64F, 0, 1)
combined_sobel = cv.bitwise_or(sobelx, sobely)

plt.imshow(sobelx)
plt.show()
plt.imshow(sobely)
plt.show()
plt.imshow(combined_sobel)
plt.show()

# canny - edge를 찾는데 가장 깔끔하다.
canny = cv.Canny(gray, 150, 175)
plt.imshow(canny)
plt.show()

cv.waitKey(0)
```
