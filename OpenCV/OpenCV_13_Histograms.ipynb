```
# histogram - pixel distribution을 보여준다.

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/park.jpg')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)

# Grayscale histogram
gray_hist = cv.calcHist([gray], [0], None, [256], [0,256])
                        # 사용할 이미지 (리스트로 표현)
                        # 채널 수
                        # 마스크
                        # 히스토그램 사이즈 (Bins의 개수)
                        # 픽셀 range

# 해당 이미지의 픽셀 분포를 알 수 있다.
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')          # 픽셀
plt.ylabel('# of pixels')   # pixel intensity
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()

cv.waitKey(0)
```

```
# mask를 사용한 히스토그램

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cats.jpg')

blank = np.zeros(img.shape[:2], dtype='uint8')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)

circle = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2),100,255,-1)
mask = cv.bitwise_and(gray, gray, mask=circle)
plt.imshow(mask)
plt.show()

# Grayscale histogram
gray_hist = cv.calcHist([gray], [0], mask, [256], [0,256])
                        # 사용할 이미지 : 리스트로 표현
                        # 채널
                        # 마스크
                        # 히스토그램 사이즈
                        # 픽셀 range

# 해당 이미지의 픽셀 분포를 알 수 있다.
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()

cv.waitKey(0)
```

```
# color histogram

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cats.jpg')
# img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/park.jpg')

blank = np.zeros(img.shape[:2], dtype='uint8')

mask = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2),100,255,-1)
masked = cv.bitwise_and(img, img, mask=mask)
plt.imshow(masked)
# plt.show()

plt.figure()
plt.title('Color Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')

# color histogram
colors = ('b','g','r')
for i , col in enumerate(colors) :
    # hist = cv.calcHist([img], [i], None, [256],[0,256])
    hist = cv.calcHist([img], [i], mask, [256],[0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.show()

cv.waitKey(0)
```
