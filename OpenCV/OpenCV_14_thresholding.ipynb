```
# thresholing : 
# binarizing : 0과 1

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cats.jpg')

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()

# 1. Simple thresholding
threshold, thresh = cv.threshold(gray, 200, 255, cv.THRESH_BINARY)
                             # threshold 경계값, # 150이상은 255 # 이진화
plt.imshow(thresh)
plt.show()

# 위에서 한 것의 inverse
threshold, thresh_inv = cv.threshold(gray, 200, 255, cv.THRESH_BINARY_INV)
                             # threshold 경계값, # 150이상은 255 # 이진화
plt.imshow(thresh_inv)
plt.show()

#2. Adaptive Threshold
# adaptive = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 3)
adaptive = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY_INV, 13, 1)
adaptive = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV, 13, 1)
plt.imshow(adaptive)
plt.show()

cv.waitKey(0)
```
