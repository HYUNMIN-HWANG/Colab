```
# AND OR XOR

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

blank = np.zeros((400,400), dtype='uint8')

rectangle = cv.rectangle(blank.copy(), (30,30),(370,370),255, -1)
circle = cv.circle(blank.copy(), (200,200),200,255,-1)

plt.imshow(rectangle)
plt.show()
plt.imshow(circle)
plt.show()

# bitwise AND -> intersecting regions 공통인 부분을 black으로 변환
bitwise_and = cv.bitwise_and(rectangle, circle) 
plt.imshow(bitwise_and)
plt.show()

# bitwise OR -> non-intersecting & intersecting regions 둘 중 하나라도 있는 부분을 black으로 변환
bitwise_or = cv.bitwise_or(rectangle, circle)   
plt.imshow(bitwise_or)
plt.show()

# bitwise XOR -> non-intersecting region
bitwise_xor = cv.bitwise_xor(rectangle, circle)
plt.imshow(bitwise_xor)
plt.show()

# bitwise NOT -> 반전시킴
bitwise_not1 = cv.bitwise_not(rectangle)
bitwise_not2 = cv.bitwise_not(circle)
plt.imshow(bitwise_not1)
plt.show
plt.imshow(bitwise_not2)
plt.show()
# img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cats.jpg')

cv.waitKey(0)
```
