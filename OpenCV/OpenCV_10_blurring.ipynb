```
# smothing, blurring
# 용어
# kernel size : number of cloumns & rows

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cats.jpg')

#BGR -> RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)    # plt는 BGR인식을 못함 RGB로 바꿔줘야 한다.
plt.imshow(rgb)
plt.show()

# [1] Averaging
# kernel를 중심으로 주변에 것들을 평균화한다. (평균하면 다 중앙값이 됨)
average = cv.blur(rgb, (3,3))
plt.imshow(average)
plt.show()

# [2] Gaussian Blur
# 더 자연스러운 blur
gauss = cv.GaussianBlur(rgb, (3,3), 0)
plt.imshow(gauss)
plt.show()

# [3] Median Blur
# Median 값을 찾는다.
median = cv.medianBlur(rgb, 3)
plt.imshow(median)
plt.show()

# [4] Bilatera Blur
# computer vision에서 많이 사용됨
bilateral = cv.bilateralFilter(rgb, 10, 35, 25)
plt.imshow(bilateral)
plt.show()

cv.waitKey(0)
```
