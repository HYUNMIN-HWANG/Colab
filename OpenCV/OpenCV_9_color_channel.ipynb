```
# splt color cahnnel
# Blue, Green, Red로 분리한다.

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/park.jpg')

blank = np.zeros(img.shape[:2], dtype='uint8')

#BGR -> RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)    # plt는 BGR인식을 못함 RGB로 바꿔줘야 한다.
plt.imshow(rgb)
plt.show()

r,g,b = cv.split(rgb)

blue = cv.merge([blank, blank, b])  # 한 색깔로만 구성
green = cv.merge([blank, g, blank])
red = cv.merge([r, blank, blank])

# plt.imshow(b)
plt.imshow(blue)
plt.show()
# plt.imshow(g)
plt.imshow(green)
plt.show()
# plt.imshow(r)
plt.imshow(red)
plt.show()

print(rgb.shape)
print(r.shape)
print(g.shape)
print(b.shape)

# (427, 640, 3)
# (427, 640)    # 한 색깔로만 구성되어 있기 때문에 세번째 인자가 없다.
# (427, 640)
# (427, 640)

merge = cv.merge([r,g,b])
plt.imshow(merge)
plt.show()

cv.waitKey(0)
```
