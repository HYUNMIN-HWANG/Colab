```
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/lady.jpg')
# img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/group 2.jpg')
img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/group 1.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(gray)
plt.show()

# 사람 신체를 수분해둔 파일
haar_cascade = cv.CascadeClassifier('./MyDrive/MyDrive/Colab/OpenCV/haar_face.xml')

# 얼굴 탐지
faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)
                                                                  # 민감성 조절
# 탐지한 얼굴의 개수
print(f'Nmber of faces found = {len(faces_rect)}')

# 얼굴이 있는 부분에 사각형으로 표시한다.
for (x,y,w,h) in faces_rect :
    cv.rectangle(img, (x,y), (x+w, y+h), (0,255,0), thickness=2)

plt.imshow(img)
cv.waitKey(0)
```
