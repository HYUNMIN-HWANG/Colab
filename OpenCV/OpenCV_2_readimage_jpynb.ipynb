```
import cv2 as cv
import matplotlib.pyplot as plt

# read videos
capture = cv.VideoCapture('./MyDrive/MyDrive/Colab/OpenCV/dog.mp4')
# display video
while True :
    isTure, frame = capture.read()  # 동영상을 프레임 by 프레임으로 읽는다.
    cv.imshow('Video', frame)
    if cv.waitKey(20) & 0xFF == ord('d'):   # stop video
        break

capture.release()
cv.destroyAllWindows()
```
