```
# Masking
# 이미지 중 포커스하고 싶은 부분에 포커스 한다.

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cats.jpg')
blank = np.zeros(img.shape[:2], dtype='uint8')
plt.imshow(blank)

# mask = cv.circle(blank, (img.shape[1]//2, img.shape[0]//2 + 45), 100, 255, -1)
circle = cv.circle(blank.copy(), (img.shape[1]//2+45, img.shape[0]//2), 100, 255, -1)
plt.imshow(mask)
plt.show()


# 가운데만 원으로 뚫어놓고 나머지는 가린다.
masked = cv.bitwise_and(img, img, mask=mask)
plt.imshow(masked)
plt.show()

rectangle = cv.rectangle(blank.copy(), (30,30),(370,370), 255, -1)
weired_shape = cv.bitwise_and(circle, rectangle)
plt.imshow(weired_shape)
plt.show()

masked_weired = cv.bitwise_and(img, img, mask=weired_shape)
plt.imshow(masked_weired)
plt.show()

cv.waitKey(0)
```
