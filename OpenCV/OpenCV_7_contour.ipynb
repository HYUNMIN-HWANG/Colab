```
# contour : edge보다 물체 인식을 하는 데 더 유용하다.

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cats.jpg')

# Gray색으로 변환
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray)

blank = np.zeros(img.shape, dtype='uint8')
plt.imshow(blank)

# # blur
blur = cv.GaussianBlur(gray, (5,5), cv.BORDER_DEFAULT)
plt.imshow(blur)    # >> blur 한 후 contour하면 좀 더 요약된 선들만 보인다.

# [1] Canny
canny = cv.Canny(blur, 125, 175)
plt.imshow(canny)

# [2] threshold
ret, thresh = cv.threshold(canny, 125, 255, cv.THRESH_BINARY)
                        # 이진화 : 125를 기준으로 이하면 0, 이상이면 255 
plt.imshow(thresh)

# contour
contours, hierarchies = cv.findContours(thresh, cv.RETR_LIST, cv.CHAIN_APPROX_NONE)
                                            # return all the contour    # how approximate
print(f'{len(contours)} contour(s) found!') # 2794 --> 380 --> 839

# draw contour  > 검은색 배경에 contour 선을 그린다.
cv.drawContours(blank, contours, -1, (0,0,255), 2)
plt.imshow(blank)

cv.waitKey(0)
```
