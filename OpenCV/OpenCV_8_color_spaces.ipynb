```
# change color spaces 
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/park.jpg')

# plt는 BGR이 없다. RGB로 바꿔서 인식한다.
plt.imshow(img)
plt.show()

# BGR -> Grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img)
plt.show()

# BGR -> HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)    # green, red
plt.imshow(hsv)
plt.show()

# BGR -> LAB
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)    # wash down version
plt.imshow(lab)
plt.show()

#BGR -> RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)    # plt는 BGR인식을 못함 RGB로 바꿔줘야 한다.
plt.imshow(rgb)
plt.show()

# 역 방향으로 변환하는 것도 가능하다.
# HSC -> BGR
hsc_bgr = cv.cvtColor(hsv, cv.COLOR_HSV2BGR)
plt.imshow(hsc_bgr)
plt.show()

# lab -> BGR
lab_bgr = cv.cvtColor(lab, cv.COLOR_LAB2BGR)
plt.imshow(lab_bgr)
plt.show()

# cv.waitKey(0)
```
