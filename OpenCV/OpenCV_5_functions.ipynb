```
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/cat.jpg')
img = cv.imread('./MyDrive/MyDrive/Colab/OpenCV/park.jpg')
# plt.imshow(img)

# 1 Converting to Grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
# plt.imshow(gray)

# 2 Blur image (noise)
blur = cv.GaussianBlur(img, (7,7), cv.BORDER_DEFAULT)
                            # blur 정도 (반드시 홀수)
# plt.imshow(blur)

# 3 Edge Cascade
canny = cv.Canny(blur, 125, 175)
plt.imshow(canny)

# 4 Dilating the image
dilated = cv.dilate(canny, (7,7),iterations=3)
plt.imshow(dilated)

# 5 Eroding
eroded = cv.erode(dilated, (3,3), iterations=3)
plt.imshow(eroded)

# 6 Resize
resize = cv.resize(img, (500,500), interpolation=cv.INTER_AREA)  # 500, 500으로 크기 변환
plt.imshow(resize)

# 7 Crop
crop = img[50:200, 200:400]
plt.imshow(crop)

cv.waitKey(0)
```
